```python
import pandas as pd

df = pd.read_csv(r'D:\github\Customer_shopping_behaviour_analysis\data\customer_shopping_behavior.csv')

df.head()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
df.to_csv(
    r'D:\github\Customer_shopping_behaviour_analysis\data\customer_shopping_behavior_cleaned.csv',
    index=False
)

print("Cleaned dataset saved!")
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("customer.db")

pd.read_sql_query("SELECT * FROM customer LIMIT 5", conn)
```

```python
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
```

```python
pd.read_sql_query("SELECT * FROM customer LIMIT 5;", conn)
```

```python
pd.read_sql_query("""
SELECT gender, SUM(purchase_amount) as revenue
FROM customer
GROUP BY gender
""", conn)
```

```python
pd.read_sql_query("PRAGMA table_info(customer);", conn)
```

```python
pd.read_sql_query("""
SELECT Gender,
       SUM([Purchase Amount (USD)]) AS revenue
FROM customer
GROUP BY Gender
""", conn)
```

```python
pd.read_sql_query("""
SELECT Customer ID, [Purchase Amount (USD)]
FROM customer
WHERE [Discount Applied] = 'Yes'
  AND [Purchase Amount (USD)] >= (
      SELECT AVG([Purchase Amount (USD)]) FROM customer
  )
""", conn)
```

```python
pd.read_sql_query("""
SELECT [Customer ID], [Purchase Amount (USD)]
FROM customer
WHERE [Discount Applied] = 'Yes'
  AND [Purchase Amount (USD)] >= (
      SELECT AVG([Purchase Amount (USD)]) FROM customer
  )
""", conn)
```

```python
pd.read_sql_query("""
SELECT [Item Purchased],
       ROUND(AVG([Review Rating]),2) AS avg_rating
FROM customer
GROUP BY [Item Purchased]
ORDER BY avg_rating DESC
LIMIT 5
""", conn)
```

```python
pd.read_sql_query("""
SELECT [Shipping Type],
       ROUND(AVG([Purchase Amount (USD)]),2) AS avg_spend
FROM customer
GROUP BY [Shipping Type]
""", conn)
```

```python
pd.read_sql_query("""
SELECT [Subscription Status],
       COUNT([Customer ID]) AS total_customers,
       ROUND(AVG([Purchase Amount (USD)]),2) AS avg_spend,
       ROUND(SUM([Purchase Amount (USD)]),2) AS total_revenue
FROM customer
GROUP BY [Subscription Status]
""", conn)
```

```python
pd.read_sql_query("""
SELECT [Item Purchased],
       ROUND(100.0 * SUM(CASE WHEN [Discount Applied] = 'Yes' THEN 1 ELSE 0 END)
       / COUNT(*),2) AS discount_rate
FROM customer
GROUP BY [Item Purchased]
ORDER BY discount_rate DESC
LIMIT 5
""", conn)
```

```python
pd.read_sql_query("""
SELECT 
CASE 
    WHEN [Previous Purchases] = 1 THEN 'New'
    WHEN [Previous Purchases] BETWEEN 2 AND 10 THEN 'Returning'
    ELSE 'Loyal'
END AS customer_segment,
COUNT([Customer ID]) AS total_customers
FROM customer
GROUP BY customer_segment
""", conn)
```

```python
pd.read_sql_query("""
SELECT *
FROM (
    SELECT [Category],
           [Item Purchased],
           COUNT(*) AS total_orders,
           ROW_NUMBER() OVER (
               PARTITION BY [Category]
               ORDER BY COUNT(*) DESC
           ) AS rank
    FROM customer
    GROUP BY [Category], [Item Purchased]
)
WHERE rank <= 3
""", conn)
```

```python
pd.read_sql_query("""
SELECT [Subscription Status],
       COUNT([Customer ID]) AS repeat_buyers
FROM customer
WHERE [Previous Purchases] > 5
GROUP BY [Subscription Status]
""", conn)
```

```python
pd.read_sql_query("""
SELECT Age,
       SUM([Purchase Amount (USD)]) AS revenue
FROM customer
GROUP BY Age
ORDER BY revenue DESC
""", conn)
```
